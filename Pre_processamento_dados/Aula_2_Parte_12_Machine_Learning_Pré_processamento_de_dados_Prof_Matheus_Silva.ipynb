```python
import pandas as pd
df = pd.DataFrame({
    'atributo1': [1,2,3],
    'atributo2': [5,6,7],
    'rotulo': ['alto', 'medio', 'baixo']
})
df
```

```python
from sklearn.preprocessing import LabelBinarizer
help(LabelBinarizer)
```

```python
df
```

```python
atributos = df.drop('rotulo', axis=1)
atributos
```

```python
rotulo = df['rotulo']
rotulo
```

```python
l_binarizer = LabelBinarizer()
l_binarizer
```

```python
l_binarizer.fit(rotulo)
```

```python
l_binarizer.classes_
```

```python
l_binarizer.transform(rotulo)
```

```python
import numpy as np
l_binarizer.inverse_transform(np.array([
    [0,0,1],
    [0,0,1],
    [0,1,0],
    [0,1,0],
    [1,0,0],
    [1,0,0],
]))
```
