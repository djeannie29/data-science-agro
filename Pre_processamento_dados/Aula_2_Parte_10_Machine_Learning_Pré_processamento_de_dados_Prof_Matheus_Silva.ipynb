# Imputers


```python
import pandas as pd
df = pd.DataFrame({
    'idade': [14,25,23,31,31,None,18,20,15,17,60]
})
df
```

```python
from sklearn.impute import SimpleImputer
help(SimpleImputer)
```

```python
constant_imputer = SimpleImputer(strategy='constant', fill_value=-1)
constant_imputer
```

```python
df
```

```python
array = df['idade'].to_numpy().reshape(-1,1)
array
```

```python
constant_imputer.fit(array) #treinamento
```

```python
df['constant'] = constant_imputer.transform(array)
df
```

```python
df_pandas = pd.DataFrame({
    'idade': df['idade'].copy()
})
constant = -1
df_pandas[df_pandas.isnull()] = constant
df_pandas
```

```python
mean_imputer = SimpleImputer(strategy='mean')
mean_imputer
```

```python
mean_imputer.fit(array) #treinamento
```

```python
mean_imputer.transform(array)
```

```python
df['mean'] = mean_imputer.transform(array)
df
```

```python
df_pandas['idade'].mean()
```

```python
df_pandas = pd.DataFrame({
    'idade': df['idade'].copy()
})
media = df_pandas['idade'].mean()
df_pandas[df_pandas.isnull()] = media
df_pandas
```

```python
median_imputer = SimpleImputer(strategy='median')
median_imputer
```

```python
median_imputer.fit(array) #treinamento
```

```python
median_imputer.transform(array)
```

```python
df['median'] = median_imputer.transform(array)
df
```

```python
df_pandas = pd.DataFrame({
    'idade': df['idade'].copy()
})
mediana = df_pandas['idade'].median()
df_pandas[df_pandas.isnull()] = mediana
df_pandas
```

# Encoder



```python
import pandas as pd
dados = [
    'professor',
    'medico',
    'advogado',
    'medico',
    'professor',
    'policial',
    'engenheiro',
    'engenheiro',
    'policial',
    'advogado',
]

df = pd.DataFrame({
    'profissao': dados
})
df
```

```python
#!pip install category_encoders
```

```python
from category_encoders import OrdinalEncoder
help(OrdinalEncoder)
```

```python
ordinal = OrdinalEncoder(cols='profissao')
ordinal
```

```python
ordinal.fit(df)
```

```python
import numpy as np
df['nulos'] = np.nan
df
```

```python
df = ordinal.transform(df)
df
```

```python
df_pandas = pd.DataFrame({
    'profissao': dados.copy(),
    'nulos': np.nan
})
mapa = {
    'professor':1,
    'medico':2,
    'advogado':3,
    'policial':4,
    'engenheiro':5
}
df_pandas['profissao'] = df_pandas['profissao'].map(mapa)
df_pandas
```

```python
from category_encoders import OneHotEncoder
help(OneHotEncoder)
```

```python
df = pd.DataFrame({
    'profissao': dados,
    'nulos': np.nan
})
df
```

```python
#one_hot = OneHotEncoder(cols=['profissao'])
one_hot = OneHotEncoder(cols=['profissao'], use_cat_names=True)
one_hot
```

```python
one_hot.fit(df) #treinamento
```

```python
one_hot.transform(df)
```

```python
df['profissao'].to_frame()
```

```python
one_hot.transform(df).drop('nulos', axis=1)
```

```python
pd.concat([
    df['profissao'].to_frame(),
    one_hot.transform(df).drop('nulos', axis=1)
], axis=1)
```

```python
help(pd.get_dummies)
```

```python
df_pandas = pd.DataFrame({
    'profissao': dados
})
df_pandas
```

```python
pd.concat([
    df_pandas,
    pd.get_dummies(df_pandas, columns=['profissao'])
], axis=1)
```

```python
from category_encoders import BinaryEncoder
help(BinaryEncoder)
```

```python
df = pd.DataFrame({
    'profissao': dados
})
df
```

```python
binary = BinaryEncoder(cols=['profissao'])
binary
```

```python
binary.fit(df) #treinamento
```

```python
binary.transform(df)
```

```python
pd.concat([
    df,
    binary.transform(df)
], axis=1)
```

```python
df_binary = binary.transform(df)
df_binary
```

```python
binary.inverse_transform(df_binary)
```

```python
pd.concat([
    df,
    binary.transform(df),
    binary.inverse_transform(df_binary)
], axis=1)
```
