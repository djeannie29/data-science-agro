# Condutividade Elétrica da Fazenda

A condutividade elétrica é muito utilizada como indicativo de água e nutrientes no solo. Baseado nessa análise, vamos facilitar a visualização das áreas mais propícias para plantio.

Dicionário de dados:
- Long: Longitude
- Lat: Latitude
- X: Coordenadas métricas
- Y: Coordenadas métricas
- ce1: Medição da condutividade elétrica a 30 cm de profundidade
- ce2: Medição da condutividade elétrica a 90 cm de profundidade
- alt: Altimetria, altura utilizando como base o nível do mar


Leitura do Excel

```python
import pandas as pd
df = pd.read_excel('/content/veris rio pardo_CALC.xlsx')
df
```

```python
df = df.iloc[:-6,:]
df
```

```python
df.info()
```

```python
df['Y'] = df['Y'].astype(float)
df.info()
```

```python
df.describe().apply(lambda s: s.apply('(0:.5f)'.format))
```

```python
df['ce2'].median() + 1.5*df['ce2'].std()
```

```python
mask_ce1 = ((df['ce1'] <= 0) | (df['ce1'] > 5))
mask_ce2 = ((df['ce2'] <= 0) | (df['ce2'] > 5))
inconsistentes = mask_ce1 | mask_ce2
inconsistentes
```

```python
inconsistentes.value_counts()
```

```python
df[inconsistentes]
```

```python
df = df[~inconsistentes]
df
```

```python
import seaborn as sns
paleta = sns.color_palette('RdYlGn', as_cmap=True)
# sns.palplot(paleta)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=[20,10])
sns.scatterplot(
    data=df,
    x='X',
    y='Y',
    hue='ce1',
    palette=paleta
);
```
