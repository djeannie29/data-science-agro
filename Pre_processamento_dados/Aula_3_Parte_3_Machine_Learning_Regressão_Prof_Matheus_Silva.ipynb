```python
import pandas as pd
df = pd.DataFrame({
    'cerveja': [2,3],
    'total': [20,30]
})
df
```

```python
df.plot(
    kind='scatter',
    x='cerveja',
    y='total',
    ylim=[0,60],
    xlim=[0,5],
    grid=True,
    s=200,
    title='Pre√ßo da cerveja'
);
```

```python
df
```

```python
atributo = df['cerveja'].to_frame()
rotulo = df['total'].to_frame()
atributo
```

```python
rotulo
```

```python
from sklearn.linear_model import LinearRegression
help(LinearRegression)
```

```python
modelo = LinearRegression()
modelo
```

```python
modelo.fit(atributo, rotulo)
```

```python
novos_dados = pd.DataFrame({
  'cerveja': [1,4,5]
})
novos_dados
```

```python
modelo.predict(novos_dados)
```

```python
novos_dados['total'] = modelo.predict(novos_dados)
novos_dados
```

```python
linha = pd.concat([df, novos_dados])
linha
```

```python
modelo.coef_, modelo.intercept_.round(2)
```
