```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
amostras = 1000
datas = pd.date_range('2020-01-01', '2020-01-31')
sexo = ['M', 'F']
estado = [
'AC', 'AL', 'AP', 'AM', 'BA', 'CE', 'DF', 'ES', 'GO',
'MA', 'MT', 'MS', 'MG', 'PA', 'PB', 'PR', 'PE', 'PI',
'RJ', 'RN', 'RS', 'RO', 'RR', 'SC', 'SP', 'SE', 'TO'
]
```

```python
len(estado)
```

```python
df = pd.DataFrame({
    'idade': np.random.randint(0, 101, amostras),
    'data': np.random.choice(datas, amostras),
    'nota': np.random.uniform(0,1001, amostras),
    'sexo': np.random.choice(sexo, amostras),
    'estado': np.random.choice(estado, amostras)

})
df
```

```python
df.info()
```

```python
df.describe(include='all')
```

Q2

```python
faltantes = df['nota'].sample(frac=.2, random_state=0).index
df.loc[faltantes, 'nota'] = np.nan
df
```

```python
df.info()
```

```python
df.loc[df['nota'].isna(), 'nota'] = 0
df
```

```python
df.info()
```

```python
me_18 = df['idade'] < 18
ma_80 = df['idade'] > 80
filtro = me_18 | ma_80
df[filtro]
```

```python
df.drop(df[filtro].index, inplace=True)
df
```

```python
def aprovado(nota):
  if nota >= 600:
    return 'Aprovado'
  else:
    return 'Reprovado'

aprovado(600), aprovado(400)
```

```python
df['aprovado'] = df['nota'].apply(aprovado)
df
```

```python
df['dia_semana'] = df['data'].dt.weekday
df
```

```python
mapa = {
    0: 'segunda',
    1: 'terÃ§a',
    2: 'quarta',
    3: 'quinta',
    4: 'sexta',
    5: 'sabado',
    6: 'domingo',
}
df['dia_semana'] = df['dia_semana'].map(mapa)
df
```
