# Limpeza de atributos

```python
def limpeza_outliers(df):
  dataset: df.copy()
  mask_rooms = df['mean_rooms'] > 60
  mask_bedrooms = df['mean_bedrooms'] > 25
  mask_population =  df['mean_bedrooms'] > 100
  outliers = mask_rooms | mask_bedrooms | mask_population
  return df[~outliers]
```

```python
todos_atributos = df.drop('median_house_value', axis='columns')
selec_atributos = df[['mean_rooms', 'median_income', 'housing_median_age', 'latitude']]
rotulo = df['median_house_value'].to_frame()
rotulo
```

```python
def selecionar_atributos(df, filtro=False):
  dataset = df.copy()
  if filtro:
    colunas = ['mean_rooms', 'median_income', 'housing_median_age', 'latitude']
    return dataset[colunas]
  else:
    return dataset.drop('median_house_value', axis='columns', errors='ignore')
```

```python
rotulo = df['median_house_value'].to_frame()
rotulo
```

```python
from sklearn.preprocessing import FunctionTransformer
criar_novos_campos = FunctionTransformer(geracao_atributos)
remover_outliers = FunctionTransformer(limpeza_outliers)
todos_atributos = FunctionTransformer(selecionar_atributos)
selec_atributos = FunctionTransformer(selecionar_atributos, kw_args={'filtro':True})
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import RobustScaler
transformador_todos = Pipeline([
    ('Criar Novos Campos', criar_novos_campos),
    # ('Remover Outliers', remover_outliers'),
    ('Remover Outliers', remover_outliers),
    ('Selecionar Campos', selec_atributos),
    ('Normalizar os dados', RobustScaler())
])

transformador_selec = Pipeline([
    ('Criar Novos Campos', criar_novos_campos),
    # ('Remover Outliers', remover_outliers'),
    ('Remover Outliers', remover_outliers),
    ('Selecionar Campos', selec_atributos),
    ('Normalizar os dados', RobustScaler())
])
```

```python
from time import time
resultados = {
    'nome': ['regressao selecionadas'],
    'RMSE_Treino': [],
    'MAE_Treino': []
}
for modelo in modelos:
  print()
  start = time()
  # Treino
  nome = modelo.steps[-1][0] #nome do modelo
  print(f'Nome: {nome}')
  resultados['nome'].append(nome)

  regressao.fit(df, rotulo) #treinamento
  predicao = regressao.predict(df) #previsão
  metrica = mse(rotulo, predicao, squared=False)
  print(f'RMSE: {metrica}')
  resultados['RMSE_Treino'].append(metrica)

  metrica = mae(rotulo, predicao)
  print(f'MAE: {metrica}')
  resultados['MAE_Treino'].append(metrica)

  stop = time()
  tempo = stop - start #Cronômetro
  print(f'Tempo Treinamento: {tempo}')
  resultados['tempo'].append(tempo)
```

```python
resultados = pd.DataFrame(resultados)
resultados
```
