```python
lista = [
    'maisena',
    'farinha',
    'açucar',
    'ovos',
    'manteiga',
    'castanha'
]
ingredientes = pd.Series(lista, name='ingredientes')
ingredientes
```

```python
dados = [200, 250, 100, 2, 150, 180]
indice = ingredientes.copy()
qtd = pd.Series(data=dados, index=indice, name='quantidade')
qtd
```

```python
df = pd.DataFrame(data={
    "ingredientes": ingredientes,
    "qtd": qtd.values
})
df
```

```python
mi_200 = df['qtd'] <= 200
mi_200
```

```python
df[mi_200]
```

```python
lista_especial = ['castanha','manteiga','leite','pão','farinha']
dentro_le = df['ingredientes'].isin(lista_especial)
df[dentro_le]
```

```python
filtro = dentro_le & mi_200
df[filtro]
```

```python
filtro = dentro_le | mi_200
df[filtro]
```

```python
~dentro_le
df[~dentro_le]
```
