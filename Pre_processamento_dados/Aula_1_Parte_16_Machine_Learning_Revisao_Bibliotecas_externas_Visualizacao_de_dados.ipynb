```python
!pip install matplotlib
import matplotlib.pyplot as plt
import pandas as pd

mes = ['jan', 'fev', 'mar', 'abr', 'mai', 'jun', 'jul', 'ago']
temperaturas = [29,26,25,20,18,17,16,18]

df = pd.DataFrame({
    'mes': mes,
    'temperatura': temperaturas
})
df
```

```python
df[::-1].plot(
    x="mes",
    y="temperatura",
    kind="bar",
    yticks=[10,20,30],
    figsize=[15,8],
    title="Temperatura DF",
    grid=True,
    legend=True,
    ylabel="°C",
    xlabel="Mês",
    ylim=[0,40],
    color="red"
);
```
