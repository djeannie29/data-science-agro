```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

modelos = [
    LogisticRegression(),
    #KNeighborsClassifier(),
    #SVC(kernel=),
    DecisionTreeClassifier(),
    RandomForestClassifier()
]

modelos
```

```python
type(modelo).__name__
```

```python
from time import time
resultado = {
    'nome':[],
    'tempo':[],
    'acuracia_treino':[],
    'acuracia_validacao':[],
    'f1_treino':[],
    'f1_validacao':[]
}
for modelo in modelos:
  nome = type(modelo).__name__
  resultado['nome'].append(nome)
  start = time()

  # Treinamento
  modelo.fit(x_train, y_train)
  predicoes_train = modelo.predict(x_train)
  metrica = acc(y_train, predicoes_train)
  resultado['acuracia_treino'].append(metrica)
  metrica = f1(y_train, predicoes_train, average='weighted')
  resultado['f1_treino'].append(metrica)

  # Treinamento
  # modelo.fit(x_val, y_val) to dataset de teste e valisação o modelo nã pode treinar com novamente os dados
  modelo.fit(x_val, y_val)
  predicoes_val = modelo.predict(x_val)
  metrica = acc(y_val, predicoes_val)
  resultado['acuracia_validacao'].append(metrica)
  metrica = f1(y_val, predicoes_val, average='weighted')
  resultado['f1_validacao'].append(metrica)

  stop = time()
  tempo = stop-start
  resultado['tempo'].append(tempo)
```

```python
resultado
```

```python
resultado = pd.DataFrame(resultado)
resultado
```
